```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

pd.pandas.set_option('display.max_columns',None)
```

```python
df=pd.read_csv("Dataset/training_set.csv")
df.head()
```

```python
df = df.drop('Unnamed: 0',axis='columns')
df.head(2)
```

```python
df.shape
```

```python
df.Y.unique()
```

# Initial Observation 

There are 57 independent variables (X1:X57) or features.

There is 1 target variable, Y.

All the features are numerical.

The target variable denotes the class of the instance. 

There are two classes - 0 and 1. So, it is a binomial classification problem.

# Class Distribution (Balance of Dataset)

```python
class_0 = len(df[df.Y==0])
class_1 = len(df[df.Y==1])
print ("Ratio of Class 0 to Class 1 is %.2f:1"%(class_0/class_1))
```

```python
classes = ['Class 0','Class 1']
num_instances = [class_0,class_1]
plt.bar(classes, num_instances, color = ['red','orange'])
plt.xlabel("Class")
plt.ylabel("No. of instances")
plt.title("Class Distribution")
plt.show()
```

# Missing Values

```python
features_with_na = [feature for feature in df.columns if df[feature].isnull().sum()>=1]

for feature in features_with_na:
    print (feature, " has ", round(df[feature].isnull().mean()*100,2),"% missing values")
```

As we can see, there are no features with null values.

# Numerical Variables

All the features are numerical variables.

```python
numerical_features = [feature for feature in df.columns if feature != 'Y']
```

Now, there can be two types of numerical features - discrete features and continuous features.

Let us inspect the two seperately.

### Discrete Features 

```python
discrete_features = [feature for feature in numerical_features if len(df[feature].unique())<=20]

print ("Number of Discrete Features ", len(discrete_features))
```

There are no discrete features, only continuous variables.

### Continuous Features 

```python
continuous_features = numerical_features
```

#### Check Distribution 

```python
for i,feature in enumerate(continuous_features):
    data = df.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

All the features are completely skewed towards zero.

#### Check the dependence of Target Variable on each feature 

```python
for i,feature in enumerate(continuous_features):
    sns.barplot(df['Y'],df[feature])
    plt.xlabel('Y')
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```

Each barplot gives a very good idea about how the feature and the target variables are related. 


Let us consider the last plot - X57 vs Y.

The plot shows that when X57 has lower value (around 150), Y tends to be 0.
And X57 tends to have a much higher value when Y = 1. However, this value varies a lot which is evident from the black CI line at the top.

#### Check Central Tendency and Range 

Let us check the mean, median to find the central tendency of each column. Along with that let us also check the min and max value which will give us an idea of the range of that column.

```python
df.describe()
```

```python
print ("Median Values")
for feature in continuous_features:
    print (feature, round(df[feature].median(),2))
```

For most of them, the median is 0 and mean is around 0. So, we can assume that the values are mostly 0.

```python
for feature in continuous_features:
    print (feature, 'has ', round(len(df[df[feature]==0.0])/len(df)*100,2), '%  0 values')
```

#### Check Outliers for a few variables (Will do proper Outlier Detection in Feature Engineering notebook) 

We can use log function to transform a skewed distribution to normal distribution. But log(0) is undefined. So, we can try this on the 3 X variables that do not have 0.

```python
### Outliers

for feature in continuous_features:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

It is evident that X55 and X56 have quite a few outliers. We need to sort this out in Feature Engineering
