# Train - Validation Split

Based on the given requirements, I am dividing the total training set into train set and validation set in th ratio of 4:1

It is the best practice to make a stratified split i.e. the proportion of different Y classes should be same in the two splits.

```python
y = df[['Y']]
X = df.drop('Y',axis='columns')

X_train,X_validation,y_train,y_validation = train_test_split(X,y,train_size = 0.8, stratify=y)
X_train.shape, X_validation.shape
```

```python
validation_set = pd.concat([X_validation,y_validation],axis='columns')
validation_set.head()
```

# Outlier Detection

```python
X_train.describe()
```

For almost all the features, most of the values are 0. As a result, if we use central tendency measures or percentiles on all the features, then most of the instances will be removed as outliers. 

I am going to use IQR method on a copy of the dataset to show this.

This can happen when there are a lot of numerical variables. As a result there is a high chance for each instance, that at least one value can lie outside this IQR zone. As a result, the whole instance is removed.

```python
X_train_copy = X_train.copy()
len_before = len(X_train_copy)
for feature in X_train_copy.columns:
    q1,q3 = X_train_copy[feature].quantile([0.25,0.75])
    iqr = q3 - q1
    lower_limit = q1 - 1.5*iqr
    upper_limit = q3 + 1.5*iqr

    X_train_copy = X_train_copy[(X_train_copy[feature]<=upper_limit) & (X_train_copy[feature]>=lower_limit)]
len_after = len(X_train_copy)
print (len_before - len_after, " outliers removed")
```

If I use traditional methods, I will end up removing almost all the instances. So, instead of considering three standard deviations in Z-score, I am going to use 8 standard deviations (to remove extreme outliers)

```python
len_before = len(X_train)
for feature in X_train.columns:
    max_threshold = X_train[feature].mean() + 8*X_train[feature].std()
    min_threshold = X_train[feature].mean() - 8*X_train[feature].std()

    X_train = X_train[(X_train[feature]<=max_threshold) & (X_train[feature]>=min_threshold)]
len_after = len(X_train)
print (len_before - len_after, " outliers removed")
```

```python
len(X_train)
```

```python
y_train = y_train[y_train.index.isin(X_train.index)]
len(y_train)
```

# Min Max Normalization

```python
X_train = X_train.reset_index().drop('index',axis='columns')
y_train = y_train.reset_index().drop('index',axis='columns')
```

```python
features = X_train.columns
scaler=MinMaxScaler()
scaler.fit(X_train[features])
X_train = pd.DataFrame(scaler.transform(X_train[features]),columns=features)
X_train.describe()
```

```python
scaler_filename = "Models/MinMaxScaler.save"
joblib.dump(scaler, scaler_filename)
```

```python
train_set = pd.concat([X_train,y_train],axis='columns')
train_set.to_csv('Dataset/train_set_after_feature_engineering.csv',index=False)
validation_set.to_csv('Dataset/validation_set.csv',index=False)
```
