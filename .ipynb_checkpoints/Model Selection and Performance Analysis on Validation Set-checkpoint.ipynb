```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import time

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier

from sklearn.model_selection import GridSearchCV

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

from sklearn.preprocessing import MinMaxScaler

from sklearn.externals import joblib
import pickle

pd.pandas.set_option('display.max_columns',None)
```

```python
df=pd.read_csv("Dataset/train_set_after_feature_selection.csv")
df.head()
```

```python
y = df[['Y']]
X = df.drop('Y',axis='columns')
X.shape
```

# Model Selection using Grid Search CV

```python
model_params = {
    'KNN': {
        'model':KNeighborsClassifier(),
        'params':{
            'n_neighbors':[7,9,11,13,15],
            'weights':['uniform','distance']
        }
    },
    'Naive Bayes': {
        'model':GaussianNB(),
        'params':{
            'priors':[None,[0.5,0.5]]
        }
    },
    'Random Forest':{
        'model':RandomForestClassifier(),
        'params' : {
            'n_estimators':[50,75,100,125,150,175,200]
        }
    },
    'Logistic Regression':{
        'model':LogisticRegression(),
        'params':{
            'C':[1,0.75,0.5],
            'solver':["lbfgs",'sag','saga','newton-cg']
        }
    },
    'SVM': {
        'model':SVC(gamma="auto"),
        'params' : {
            'C':[1,10,20],
        }
    },
    'XG Boost': {
        'model':XGBClassifier(),
        'params' : {
            'n_estimators':[50,100,150],
            'max_depth':[3,4,5],
            'eval_metric':['logloss','error']
        }
    },
    'ANN': {
        'model':MLPClassifier(),
        'params' : {
            'hidden_layer_sizes':[(100,),(54,108,54,27)],
            'activation':['logistic','relu'],
            'solver':['sgd','adam']
        }
    }
}
```

I am going to use the macro-averaged F1 score since this is not a perfectly balanced dataset.

```python
scores = []

for model_name, mp in model_params.items():
    start = time.time()
    print (model_name," started...")
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False,scoring='f1_macro')
    clf.fit(X,y)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
    print (model_name," completed...")
    
    
    end = time.time()
    duration = end - start
    if (duration>=3600):
        hours = duration//3600
        mins = (duration%3600)//60
        secs = duration%60
    elif (duration>=60):
        hours = 0
        mins= duration//60
        secs = duration%60
    else:
        hours = 0
        mins = 0
        secs = duration
    print (round(hours,2), " hr(s) ",round(mins,2)," min(s) ",round(secs,2), " sec(s)\n")
```

```python
results_df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
results_df.sort_values('best_score',ascending=False)
```

```python
print ("XG Boost Best params")
results_df.best_params[5]
```

The two best models are Random Forest and XG Boost. I am going to consider both of them for validation set since there scores are similar. 

Based on their performance on validation set, I am going to select one model and save it to test on the Hold-Out Test Set.

# Training the two models with their best parameters

```python
model_random_forest = RandomForestClassifier(n_estimators=175)
model_random_forest.fit(X,y)
```

```python
with open('Models/Initial_model_for_validation_set_Random_Forest','wb') as f:
    pickle.dump(model_random_forest,f)
```

```python
model_xg_boost = XGBClassifier(eval_metric='logloss',n_estimators=150,max_depth=3)
model_xg_boost.fit(X,y)
```

# Performance Analysis on Validation Set

```python
validation_set=pd.read_csv('Dataset/validation_set.csv')
validation_set.head()
```

```python
y_validation = validation_set.Y
X_validation = validation_set.drop('Y',axis='columns')
```

#### Min Max Normalization on Validation Set (using the scaler used on Train Set) 

```python
scaler_filename = "Models/MinMaxScaler.save"
scaler = joblib.load(scaler_filename)

features = X_validation.columns

X_validation = pd.DataFrame(scaler.transform(X_validation[features]),columns=features)
X_validation.head()
```

#### Feature Selection in Validation Set using the features selected during training 

```python
selected_features_in_training_set = X.columns.values
X_validation = X_validation[selected_features_in_training_set]
X_validation.shape
```

## Model Performance on Validation Set 

#### Model 1 - Random Forest 

```python
print ("Accuracy is ", round(model_random_forest.score(X_validation,y_validation)*100,3))
```

```python
y_predicted = model_random_forest.predict(X_validation)
```

```python
print (classification_report(y_validation, y_predicted))
```

```python
cm = confusion_matrix(y_validation, y_predicted)
ax = sns.heatmap(cm,annot=True,fmt="d",cmap='Greens')
plt.xlabel('Predicted')
plt.ylabel('Truth')
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

#### Model 2 - XG Boost 

```python
print ("Accuracy is ", round(model_xg_boost.score(X_validation,y_validation)*100,3))
```

```python
y_predicted = model_xg_boost.predict(X_validation)
print (classification_report(y_validation, y_predicted))
```

```python
cm = confusion_matrix(y_validation, y_predicted)
ax = sns.heatmap(cm,annot=True,fmt="d",cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Truth')
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

We can see that both the models are doing pretty well. 

XG Boost is slightly better than Random Forest.

From the confusion matrix, we can see that XG Boost is predicting accurately just two more instances of the validation set(one from each class).

From classification report, we can see that both the models are doing better for Class 0. It is quite expected as the the number of instances belonging to class zero is higher.

#### So, the final model chosen for training is XG Boost 

# Model Performance Analysis in terms of Various Risks

The validation split was done before any preprocessing or feature selection. As a result, there was no data leakage from validation set to train set. 

We can say that there is no overfitting. The validation score is very close to training score. 

Test Set is still not used. It will be used only for final evaluation. So, no leakage there.  

There is no sampling bias. I have used the stratify method to maintain the class distribution in both the train and validation set. 

During model selection, I have used macro-averaged F1 score since this is not a perfectly balanced dataset.

# Final training using complete training set (both validation set and partial train set)


# This model is then going to be tested on test set

I am going to use the training set initially given (before the train - validation split)

Perform all the steps that I performed, using the best models/parameters/features I obtained while training

1. Outlier Detection
2. Min Max Normalization
3. Use the features I have already selected in Feature Selection
4. Train XG Boost model with best parameters found in Grid Search

```python
train_df = pd.read_csv('Dataset/training_set.csv')
train_df = train_df.drop('Unnamed: 0',axis='columns')
train_df.head()
```

```python
y_train = train_df.Y
X_train = train_df.drop('Y',axis='columns')
```

```python
len_before = len(X_train)
for feature in X_train.columns:
    max_threshold = X_train[feature].mean() + 8*X_train[feature].std()
    min_threshold = X_train[feature].mean() - 8*X_train[feature].std()

    X_train = X_train[(X_train[feature]<=max_threshold) & (X_train[feature]>=min_threshold)]
len_after = len(X_train)
print (len_before - len_after, " outliers removed")

y_train = y_train[y_train.index.isin(X_train.index)]
```

```python
X_train = X_train.reset_index().drop('index',axis='columns')
y_train = y_train.reset_index().drop('index',axis='columns')
```

```python
features = X_train.columns
scaler=MinMaxScaler()
scaler.fit(X_train[features])
X_train = pd.DataFrame(scaler.transform(X_train[features]),columns=features)
X_train.describe()
```

```python
scaler_filename = "Models/FinalMinMaxScaler.save"
joblib.dump(scaler, scaler_filename)
```

```python
X_train = X_train[selected_features_in_training_set]
```

```python
model_xg_boost_final = XGBClassifier(eval_metric='logloss',n_estimators=150,max_depth=3)
model_xg_boost_final.fit(X_train,y_train)
```

```python
with open('Models/Final_model_XGBoost','wb') as f:
    pickle.dump(model_xg_boost_final,f)
```
