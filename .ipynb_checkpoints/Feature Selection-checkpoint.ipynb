```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import mutual_info_classif

pd.pandas.set_option('display.max_columns',None)
```

```python
df=pd.read_csv("Dataset/train_set_after_feature_engineering.csv")
df.head()
```

```python
y = df[['Y']]
X = df.drop('Y',axis='columns')
```

I will try to find the most important features in the following three steps:

1. Variance Threshold: Remove columns where all values are same. (zero variance)
2. Pearson Correlation: For each pair of columns, remove one if they are highly correlated.
3. Mutual Information Gain: Find the features on which the target variable is most dependent on

# Variance Threshold

```python
var_threshold = VarianceThreshold(threshold=0.0)
var_threshold .fit(X)
print ("Number of constant columns: ",len(X.T)-sum(var_threshold.get_support()))
```

# Pearson's Correlation Coefficient

```python
def correlation(dataset,threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if corr_matrix.iloc[i,j] > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
corr_features = correlation(X,0.75)
len(set(corr_features))
```

```python
corr_features
```

```python
X = X.drop(corr_features,axis='columns')
X.head()
```

# Mutual Information Gain

```python
mutual_info = mutual_info_classif(X,y)
mutual_info = pd.Series(mutual_info)
mutual_info.index = X.columns
print ("Features with Information Gain above 0.05")
mutual_info = mutual_info[mutual_info.values>0.05]
mutual_info.sort_values(ascending=False)
```

```python
X = X[mutual_info.index]
X.head()
```

```python
train_set = pd.concat([X,y],axis='columns')
train_set.to_csv('Dataset/train_set_after_feature_selection.csv',index=False)
```
