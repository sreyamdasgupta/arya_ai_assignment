```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.externals import joblib
import pickle

pd.pandas.set_option('display.max_columns',None)
```

```python
test_set = pd.read_csv('Dataset/test_set.csv')
test_ids = test_set['Unnamed: 0']
test_set = test_set.drop('Unnamed: 0',axis='columns')
test_set.head()
```

```python
scaler_filename = "Models/FinalMinMaxScaler.save"
scaler = joblib.load(scaler_filename)

features = test_set.columns

test_set = pd.DataFrame(scaler.transform(test_set[features]),columns=features)
test_set.head()
```

```python
X = pd.read_csv('Dataset/train_set_after_feature_selection.csv')
X = X.drop('Y',axis='columns')
selected_features_in_training_set = X.columns.values
test_set = test_set[selected_features_in_training_set]
test_set.shape
```

```python
with open('Models/Final_model_XGBoost','rb') as f:
    model = pickle.load(f)
    
predictions = model.predict(test_set)
predictions
```

```python
predictions_file = pd.concat([test_ids,pd.Series(predictions)],axis='columns')
predictions_file.columns = ['ID', 'Y']
predictions_file
```

```python
predictions_file.to_csv('prediction.csv',index=False)
```
